```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import stats
from mpl_toolkits.basemap import Basemap

data=pd.read_csv("Inpatient_Prospective_Payment_System__IPPS__Provider_Summary_for_the_Top_100_Diagnosis-Related_Groups__DRG__-_FY2011.csv")
data.head()
data[[' Average Covered Charges ', ' Average Total Payments ', 'Average Medicare Payments']]=data[[' Average Covered Charges ', ' Average Total Payments ', 'Average Medicare Payments']].replace('[\$,)]','', regex=True).astype(float)
dataState=data[['Provider State', ' Total Discharges ', ' Average Covered Charges ', ' Average Total Payments ', 'Average Medicare Payments']].groupby('Provider State')

avgDischarges=dataState[' Total Discharges ']
plt.barh(np.arange(len(avgDischarges)), avgDischarges.sum(), align = 'center')
plt.yticks(np.arange(len(avgDischarges)), avgDischarges.groups)
plt.xlabel('Total Number of Discharges', fontsize=20)
plt.ylabel('Provider State', fontsize=20)
plt.title('Total Number of Discharges by State')
fig = plt.gcf()
fig.set_size_inches(13, 10)
fig.subplots_adjust(bottom=0.1, top=1)
```

```python
avgCoveredCharges=dataState[' Average Covered Charges '].mean()
plt.barh(np.arange(len(avgCoveredCharges)), avgCoveredCharges.values, align = 'center')
plt.yticks(np.arange(len(avgCoveredCharges)), avgCoveredCharges.index)
plt.xlabel('Average Covered Charges', fontsize=20)
plt.ylabel('Provider State', fontsize=20)
plt.title('Average Covered Charges by State')
fig = plt.gcf()
fig.set_size_inches(13, 10)
fig.subplots_adjust(bottom=0.1, top=1)
```

```python
avgTotalPayments=dataState[' Average Total Payments '].mean()
plt.barh(np.arange(len(avgTotalPayments)), avgTotalPayments.values, align = 'center')
plt.yticks(np.arange(len(avgTotalPayments)), avgTotalPayments.index)
plt.xlabel('Average Total Payments', fontsize=20)
plt.ylabel('Provider State', fontsize=20)
plt.title('Average Total Payments by State')
fig = plt.gcf()
fig.set_size_inches(13, 10)
fig.subplots_adjust(bottom=0.1, top=1)
```

```python
avgMedicarePayments=dataState['Average Medicare Payments'].mean()
plt.barh(np.arange(len(avgMedicarePayments)), avgMedicarePayments.values, align = 'center')
plt.yticks(np.arange(len(avgMedicarePayments)), avgMedicarePayments.index)
plt.xlabel('Average Medicare Payments', fontsize=20)
plt.ylabel('Provider State', fontsize=20)
plt.title('Average Medicare Payments by State')
fig = plt.gcf()
fig.set_size_inches(13, 10)
fig.subplots_adjust(bottom=0.1, top=1)
```

```python
#displays DRGs that have 10 highest average number of discharges

dataDRG=data[['DRG Definition', 'Provider State', ' Total Discharges ', ' Average Covered Charges ', ' Average Total Payments ', 'Average Medicare Payments']].groupby('DRG Definition')

DischargesDRG=dataDRG[' Total Discharges '].sum()
DischargesDRG=DischargesDRG.sort_values(ascending = False)
plt.barh(np.arange(len(DischargesDRG[0:11])), DischargesDRG[0:11], align = 'center')
plt.yticks(np.arange(len(DischargesDRG[0:11])), DischargesDRG[0:11].index)
plt.xlabel('Total Number of Discharges', fontsize=20)
plt.ylabel('DRG', fontsize=20)
plt.title('Total Number of Discharges by DRG')
fig = plt.gcf()
fig.set_size_inches(13, 10)
fig.subplots_adjust(bottom=0.1, top=1)
```

```python
avgCoveredChargesDRG=dataDRG[' Average Covered Charges '].mean()
avgCoveredChargesDRG=avgCoveredChargesDRG.sort_values(ascending=False)
plt.barh(np.arange(len(avgCoveredChargesDRG[0:11])), avgCoveredChargesDRG[0:11], align = 'center')
plt.yticks(np.arange(len(avgCoveredChargesDRG[0:11])), avgCoveredChargesDRG[0:11].index)
plt.xlabel('Average Covered Charges', fontsize=20)
plt.ylabel('DRG', fontsize=20)
plt.title('Average Covered Charges by DRG')
fig = plt.gcf()
fig.set_size_inches(13, 10)
fig.subplots_adjust(bottom=0.1, top=1)
```

```python
avgTotalPaymentsDRG=dataDRG[' Average Total Payments '].mean()
avgTotalPaymentsDRG=avgTotalPaymentsDRG.sort_values(ascending=False)
plt.barh(np.arange(len(avgTotalPaymentsDRG[0:11])), avgTotalPaymentsDRG[0:11], align = 'center')
plt.yticks(np.arange(len(avgTotalPaymentsDRG[0:11])), avgTotalPaymentsDRG[0:11].index)
plt.xlabel('Average Total Payments', fontsize=20)
plt.ylabel('DRG', fontsize=20)
plt.title('Average Total Payments by DRG')
fig = plt.gcf()
fig.set_size_inches(13, 10)
fig.subplots_adjust(bottom=0.1, top=1)
```

```python
avgMedicarePaymentsDRG=dataDRG['Average Medicare Payments'].mean()
avgMedicarePaymentsDRG=avgMedicarePaymentsDRG.sort_values(ascending=False)
plt.barh(np.arange(len(avgMedicarePaymentsDRG[0:11])), avgMedicarePaymentsDRG[0:11], align = 'center')
plt.yticks(np.arange(len(avgMedicarePaymentsDRG[0:11])), avgMedicarePaymentsDRG[0:11].index)
plt.xlabel('Average Medicare Payments', fontsize=20)
plt.ylabel('DRG', fontsize=20)
plt.title('Average Medicare Payments by DRG')
fig = plt.gcf()
fig.set_size_inches(13, 10)
fig.subplots_adjust(bottom=0.1, top=1)
```
